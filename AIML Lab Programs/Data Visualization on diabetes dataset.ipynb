```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/diabetes.csv')

# Display first few rows
print("First 5 rows of the dataset:")
display(df.head())
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Glucose'], df['Insulin'])
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.title('Glucose vs Insulin')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.style.use('seaborn-v0_8-darkgrid')   # Apply Seaborn dark grid style
plt.plot(df['Age'], df['BMI'], color='blue', marker='o', linestyle='--')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Age vs BMI with Style')
plt.show()
```

```python
outcome_counts = df['Outcome'].value_counts()
plt.figure(figsize=(6, 4))
outcome_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Distribution of Outcomes')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['BMI'], bins=20, edgecolor='black', color='orange')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('Distribution of BMI')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Glucose'], df['BloodPressure'], alpha=0.6, color='purple')
plt.xlabel('Glucose')
plt.ylabel('Blood Pressure')
plt.title('Glucose vs Blood Pressure (Scatter Plot)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df[['Insulin', 'BMI', 'Age']].plot(kind='area', stacked=False)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Area Plot of Insulin, BMI, and Age')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left: Age distribution
axes[0].hist(df['Age'], bins=20, edgecolor='black', color='teal')
axes[0].set_title('Distribution of Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

# Right: Pregnancies vs Outcome
axes[1].scatter(df['Pregnancies'], df['Outcome'], alpha=0.5, color='coral')
axes[1].set_title('Pregnancies vs Outcome')
axes[1].set_xlabel('Pregnancies')
axes[1].set_ylabel('Outcome')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='BMI', data=df, palette='Set2')
plt.xlabel('Outcome')
plt.ylabel('BMI')
plt.title('Box Plot of BMI by Outcome')
plt.show()
```

```python
sns.pairplot(df[['Glucose', 'BMI', 'Age', 'Outcome']], hue='Outcome', diag_kind='hist')
plt.suptitle('Pair Plot of Selected Diabetes Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Diabetes Dataset')
plt.show()
```
